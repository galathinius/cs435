This is an analysis of the dataset from the following link
https://www.kaggle.com/sherrytp/airline-delay-analysis/version/2

```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('2019.csv', nrows=100)
df.columns
```

```python
df.dtypes
```

```python
df = df.drop('Unnamed: 20', 1)
```

```python
corr = df.corr()

sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
delay_corr = corr['DEP_DELAY']
delay_corr['ARR_TIME']
```

```python
keys = delay_corr.index
keys = keys[keys != 'DEP_DELAY']
useful_columns = []

for key in keys:
    if abs(delay_corr[key]) > 0.6:
        useful_columns.append({key: delay_corr[key]})
        
        
useful_columns
```
